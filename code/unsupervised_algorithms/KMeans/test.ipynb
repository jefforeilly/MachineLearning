Import classifier, plotting and data handling libraries

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans as SKLKMeans

from k_means import KMeans as MyKMeans

sns.set()
sns.set_style(style='darkgrid')
```

Read data to pandas dataframe

```python
data = pd.read_csv('xclara.csv')
data.head(4)
```

Plot data before clustering

```python
sns.set_style(style='darkgrid')
sns.regplot(*data.columns.values, data=data, fit_reg=False)
```

Fit classifier and predict labeles from sklearn and my implementation

```python
skl_clusters = SKLKMeans(n_clusters=3).fit_predict(data)
my_clusters = MyKMeans(n_clusters=3).fit_predict(data)

data['skl_clusters'] = skl_clusters
data['my_clusters'] = my_clusters
```

Visualize data from each clustering

```python
sns.lmplot(x='V1', y='V2', hue='skl_clusters', data=data, fit_reg=False)
sns.lmplot(x='V1', y='V2', hue='my_clusters', data=data, fit_reg=False)
```
